Standardization of Numerical Variables


●	Objective: To scale numerical variables for uniformity, improving the dataset’s suitability for analytical models.

Explaination:

Explaining the concept of Standardization (Z-Score Normalization):

Standardization (Z-score normalization) transforms the data into standard value which hepls for the machine learning model to learn from the data more easily.
The standardized valued lies between 0 and 1, mean (μ) of 0 and a standard deviation (σ) of 1. All the values lies between this range.
We must remember that the data of the columns that are standardized are also equivalent the data that is not standardized.

z = (x - μ)/ σ

where, x = the original data point.
μ = the mean of the data.
σ = the standard deviation of the data.

	Standardizing the numerical columns using the formula: 
    z=x-mu/σ

```python
df_st = df       # Assign the df to new variable to not update the original data
df_st.info()
```

```python
#Standardizing the "Volume" column

means = df_st['Volume'].mean()   #Find mean
stds = df_st['Volume'].std()   #Find standard deviation

# applying formula z = (x - mu) / sigma
df_standardized = (df_st['Volume']-means)/stds

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

sns.histplot(data=df_st['Volume'], kde=True, ax=axes[0])    # Original data distribution
axes[0].set_title("Original Data Distribution of Volume")

sns.histplot(data=df_standardized, kde=True, ax=axes[1])   # Standardized data distribution
axes[1].set_title("Standardized Data Distribution of Volume")

df_st['Volume'] = df_standardized         # assigning the standardized value to the original dataset for machine learning
# print(df_st)

plt.tight_layout()
plt.show()
```

Here we can observe that only scale of the data is reduced but the distribution across the plane is the same. 

Hence we can say the data is standardized 

```python
#Standardizing the "Avg Price" column

means = df_st['Avg Price'].mean()   #Find mean
stds = df_st['Avg Price'].std()   #Find standard deviation

# applying formula z = (x - mu) / sigma
df_standardized = (df_st['Avg Price']-means)/stds

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

sns.histplot(data=df_st['Avg Price'], kde=True, ax=axes[0])    # Original data distribution
axes[0].set_title("Original Data Distribution of Avg Price")

sns.histplot(data=df_standardized, kde=True, ax=axes[1])   # Standardized data distribution
axes[1].set_title("Standardized Data Distribution of Avg Price")

df_st['Avg Price'] = df_standardized         # assigning the standardized value to the original dataset for machine learning
# print(df_st)

plt.tight_layout()
plt.show()
```

    Conversion of Categorical Data into Dummy Variables

●	Objective: To transform categorical variables into a format that can be provided to ML algorithms.

Discussing the need for converting categorical data into dummy variables : 

The most important reason to convert categorical data into numerical is that our computers don't understand categorical data, the only type of data it understands is numerical. Although it can understand after converting categorical to numerical data, we must keep in mind that our main goal is to train the model so it is best practice to avoid any unnecessary clutter in the data and stick to numerical data.

Technique used for doing this is called OneHotEncding.

Reasons to do OneHotEncoding-
* Compatibility with Algorithms: Most of the algorithms take data in numerical format so it is easily compatible.

* Allows mathematical operations: Since the data is in numerical format, we can perform add, sub, mul and etc...

* Representation of Relationships: Proper encoding and decoding can be done

* Model Performance: Increases the preformace of model since there is no need to do unnecessary convertions for understanding

```python
df_cat.info()     #this variable contains all the categorical data so we need to apply OneHotEncoding to all the columns
```

```python
#Convert the column into categorical columns

encoded_df = pd.get_dummies(df_cat, drop_first=False)
encoded_df
```

```python
encoded_df.shape
```

We can infer that we have 95 columns after applying OneHotEncoding.

Note that these columns are different than the columns which we already have.

```python
df_final = df_st.drop(df_cat, axis=1)     #dropped categorical column
```

```python
# Now add the encoded columns to the final dataframe for passing it to ML algorithms
df_final = pd.concat([df_final, encoded_df], axis = 1)
df_final
```
