#### **Interview Questions:**


**1. What is lift and why is it important in Association rules?**
**Answer**
* **Lift** is a measure used in association rule mining to evaluate the strength of a rule by considering the co-occurance of itmes relative to their individual occurances.
* **Formula** : lift = P(A and B)/P(A)*P(B)
* Importance of Lift:
  * **Lift > 1** - Indicates a strong positive association between the items basically they appear more than expected by chance.
  * **Lift = 1** - Suggest No association between the items.
  * **Lift < 1** - Indicates negative association between items
* Lift helps prioritize rules by identifying strong and meaningful relationships.

**2. What is support and Confidence. How do you calculate them?**

**Answer**

**Suppport** is a measure of the frequency of an itemset in the dataset
* **Formula** : support(A) = Number of transaction containing A/Total number of transactions.
* Indicates how often an itemset appears in the data, helping filter rare itemsets.

**Confidence** is the measure of the likelihood of an item B occuring in a transaction given that A is present (A-->B)
* **Formula** : Confidence(A-->B) = Support(A and B)/Support(A)
* Reveals the strength of the rule A-->B, expressing the proportion of transactions with A that also contains B.

**3. What are some limitations or challenges of Association rules mining?**

**Answer**
* **Scalability :** Processing large dataset can be expensive due to exponential growth of candidate itemset.
* **Interpretability :** For complex dataset it is difficult to interpret the generated rules.
* **Redundancy :** Many rules generated may be redundant or trivial, making it difficult to identify the meaningful ones.
* **Threshold Dependency :** Identifying appropriate threshold for support and and confidence may be difficult and subjective to the topic.
* **Rare Itemsets :** Association rules often ignores rare but potentially significant itemsets due to low support values.
* **Data Quality :** Noisy, missing values and inconsistencies in the dataset can impact the quality and accuracy of discovered rules.
